```python
df.HomePlanet.value_counts(dropna=False, normalize=1)
```

```python
df.HomePlanet.mode()[0]
```

```python
df.HomePlanet = df.HomePlanet.fillna(df.HomePlanet.mode()[0])
```

```python
HomePlanet_map = {
    "Earth": 1,
    "Europa": 2,
    "Mars": 3
}
df.HomePlanet = df.HomePlanet.map(HomePlanet_map)
df.info()
```

```python
df.CryoSleep.value_counts(dropna=False, normalize=1)
```

```python
df.CryoSleep = df.CryoSleep.fillna(df.CryoSleep.mode()[0])
map_conditions = {
    "False": 0,
    "True": 1
}
df.CryoSleep = df.CryoSleep.map(map_conditions)
df.CryoSleep.value_counts(normalize=1)
```

```python
df["CabinDeck"] = df.Cabin.str[0]
df["CabinNum"] = df.Cabin.str[2:-2]
df["CabinSide"] = df.Cabin.str[-1]
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.drop("Cabin", axis=1, inplace=True, errors='ignore')
```

```python
df.Destination.value_counts(dropna=False, normalize=1)
```

```python
df.Destination = df.Destination.fillna(df.Destination.mode()[0])
destination_map = {
    "TRAPPIST-1e": 1,
    "55 Cancri e": 2,
    "PSO J318.5-22": 3
}
df.Destination = df.Destination.map(destination_map)
```

```python
df.Age = df.Age.astype('float64')
```

```python
df.VIP = df.VIP.fillna(df.VIP.mode()[0]).map(map_conditions)
```

```python
df[df.VIP == 1].sample(5)
```

```python
df.RoomService = df.RoomService.astype('float64')
df.FoodCourt = df.FoodCourt.astype('float64')
df.ShoppingMall = df.ShoppingMall.astype('float64')
df.Spa = df.Spa.astype('float64')
df.VRDeck = df.VRDeck.astype('float64')
```

```python
df.info()
```

```python
df.Transported = df.Transported.map(map_conditions)
```

```python
df.CabinNum = df.CabinNum.astype('float64')
```

```python
df.Age = df.Age.astype('float64')
```

```python
df.info()
```
