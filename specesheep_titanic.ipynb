```python
def sm_band_calc(row):
    row_index = 'ShoppingMall'
    if row[row_index] >= 2000:
        return '4 2000+'
    elif row[row_index] >= 500:
        return '3 500-2000'
    elif row[row_index] > 0:
        return '2 0-500'
    else:
        return '1 0'

df["SMBand"] = df.apply(sm_band_calc, axis=1)
```

```python
df_agg_sm0 = df[(df.CryoSleep == 0)][["Transported", "SMBand", "HomePlanet"]].groupby(["SMBand", "HomePlanet"], dropna=False).agg(["count", "sum", "mean"])["Transported"]
df_agg_sm0
```

```python
def transopted_calc(row) -> int:
    result = 0
    hp, dest, cs, age = 0, 0, 0, 0
    dest0, fc0, rs0, spa0, vr0, sm0 = 0, 0, 0, 0, 0, 0

    if row.CryoSleep == 1:
        hp = df_agg_hp["mean"][row.HomePlanet]
        dest = df_agg_dest["mean"][row.Destination]
        cs = df_agg_cs["mean"][row.CabinSide]
        age = df_agg_age["mean"][row.Age_band]

        result = round((hp + dest + cs + age)/4, 4)
        
        #0.8176
        res_calc = lambda x : 1 if result >= 0.8176 else 0
        result = res_calc(result)

    elif row.CryoSleep == 0:
        
        dest0 = df_agg_dest0["mean"][row.Destination]
        fc0 = df_agg_fc0["mean"][row.FCBand][row.HomePlanet]
        rs0 = df_agg_rs0["mean"][row.RSBand][row.HomePlanet]
        spa0 = df_agg_spa0["mean"][row.SpaBand][row.HomePlanet]
        vr0 = df_agg_vr0["mean"][row.VRBand][row.HomePlanet]
        sm0 = df_agg_sm0["mean"][row.SMBand][row.HomePlanet]

        result = round((fc0 + rs0 + spa0 + vr0 + sm0 + dest0)/6, 4)

        #0.335
        res_calc = lambda x : 1 if result >= 0.335 else 0
        result = res_calc(result)

    return result

df["Transportded_test"] = df.apply(transopted_calc, axis=1)
```
